Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
- The Moscow Times — Wednesday, October 2, 2002
- The Guardian — Friday, 11.10.13
- Daily News — Thursday, 18 August 1977

```python
from datetime import datetime
newspaper_list = ['The Moscow Times', 'The Guardian', 'Daily News']
dates_list = ['Wednesday, October 2, 2002', 'Friday, 11.10.13', 'Thursday, 18 August 1977']

dates_list [0] = datetime.strptime('Wednesday, October 2, 2002', '%A, %B %#d, %Y')
dates_list [1] = datetime.strptime('Friday, 11.10.13', '%A, %d.%m.%y')
dates_list [2] = datetime.strptime('Thursday, 18 August 1977', '%A, %#d %B %Y')

my_dict = dict(zip(newspaper_list, dates_list))

print (my_dict)
print (dates_list [0], type(dates_list[0]))
```

Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date.
Даты должны вводиться в формате yyyy-mm-dd. В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
from datetime import datetime, timedelta

def date_range(start, end):
    """
    Возвращает список дат в формате yyyy-mm-dd от start до end включительно.
    """
    # Преобразуем строки в объекты date, .date для формата без времени, как в yyyy-mm-dd
    # try-except -  в случае неверного формата возвращается пустой список
    
    try:
        start_date = datetime.strptime(start, '%Y-%m-%d').date()
        end_date = datetime.strptime(end, '%Y-%m-%d').date()
    except ValueError:
        return []
        
    date_list = []
    current_date = start_date

    # Итерируемся до тех пор, пока текущая дата не превысит конечную,
    # соответственно пустой список при start_date > end_date
    
    while current_date <= end_date:   
        date_list.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)
    
    return date_list
           
# --- Пример использования ---
start = "2026-12-23"
end = "2027-01-07"
result = date_range(start, end)
print(result)

```
