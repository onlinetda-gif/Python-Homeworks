# Задание 1
Датасет для домашнего задания вы найдете в разделе “Материалы к домашнему заданию” к этому занятию.
Определите, какому фильму было выставлено больше всего оценок 5.0

```python
import pandas as pd
df_1 = pd.read_csv ('movies.csv')
df_1.head()
```

```python
df_2 = pd.read_csv ('ratings.csv')
# df_2.head()
```

```python
top_movie = df_2[df_2['rating'] == 5.0]['movieId'].value_counts().idxmax()
top_title = df_1[df_1['movieId'] == top_movie]['title'].item()
number_of_5s = df_2[df_2['rating'] == 5.0]['movieId'].value_counts()

print (top_title)
print (number_of_5s)
```

```python
joined = df_2.merge(df_1, on='movieId', how='left')
joined.head()
```

```python
check_top_movie = joined[joined['rating'] == 5.0]['title'].value_counts().idxmax()
print (check_top_movie)

counts = joined[joined['rating'] == 5.0]['title'].value_counts()
print(f"Фильм: {counts.index[0]}, Кол-во пятерок: {counts.values[0]}")
```

# Задание 2
По данным файла power.csv (находится в “Материалы к домашнему заданию”) посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity

```python
data = pd.read_csv('power.csv')
data.head ()
```

```python
f_countries = data[
    ((data['country'] == 'Lithuania') | (data['country'] == 'Latvia') | (data['country'] == 'Estonia')) 
    & ((data['year'] >= 2005 ) & (data ['year'] <= 2010)) 
    & (data ['quantity']>= 0)    
    & ((data ['category'] == 4) | (data ['category'] == 12) | (data ['category'] == 21))
    ]
f_countries.head()
```

```python
sum_qty_f_countries = f_countries ['quantity']. sum()
print (sum_qty_f_countries)
```
