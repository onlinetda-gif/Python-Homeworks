### Задание 1
Переведите содержимое файла purchase_log.txt в словарь purchases вида:{‘1840e0b9d4’: ‘Продукты’, …}

```python
import json

purchases = {} 
i=0 #счетчик для проверки

with open('purchase_log.txt','r', encoding = 'utf8') as f: 
  for i, line in enumerate(f): 

    dict_ = json.loads(line) 

    key = dict_['user_id'] 
    value = dict_['category']

    purchases[key] = value

    i+=1  #для проверки, для полного реззультата надо закомментировать строки до break включительно
    if i>10:
      print (purchases)
      break
```

### Задание 2 (необязательное)
Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки. Если покупка была, сам файл visit_log.csv изменять не надо.
Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
Учтите условия для данных:
- содержимое purchase_log.txt помещается в оперативную память компьютера;
- содержимое visit_log.csv — нет, используйте только построчную обработку этого файла.

```python
with open('visit_log.csv', 'r', encoding = 'utf8') as f_visits, open('funnel.csv','w', encoding='utf-8-sig') as f_funnel:
  
    for i, line in enumerate(f_visits):
    line = line.strip().split(',')
    
    if line[0] in purchases:
        line.append(purchases[line[0]])
        ad_line = ','.join(line)

        f_funnel.write(ad_line + '\n')

        print (ad_line) # проверка того, что должно быть в файле 'funnel.csv'
```
